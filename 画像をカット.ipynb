```python
import cv2
filename = 'No1_10_color.JPG'
im = cv2.imread('./' + filename)
```

```python
# im[top : bottom, left : right]   1024/64=16, 768/64=12
# 切り出し、保存

for i in range(12):
  for j in range(16):
    im_cut = im[64*i : 64*(i+1), 64*j: 64*(j+1)]
    cv2.imwrite("./cutimage/"+filename[:-4]+str(i)+"_"+str(j)+".jpg",im_cut)
```
